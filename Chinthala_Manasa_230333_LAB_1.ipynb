```python
import numpy as np

# WGS-84 parameters
a = 6378137.0                      # semi-major axis (m)
f = 1 / 298.257223563              # flattening
e2 = 2*f - f*f
def dms_to_deg(d, m, s):
    return d + m/60 + s/3600
# Decimal Degrees to DMS
def deg_to_dms(deg):
  d = int(deg)
  m = int((deg - d) * 60)
  s = (deg - d - m/60) * 3600
  return f"{d}°{m}′{s:.4f}″"
#kanpur coordinates
lat_k = dms_to_deg(26, 26, 59.7228)    # degrees
lon_k = dms_to_deg(80, 19, 54.7356)    # degrees
h_k   = 126.630                        # meters
#hyderabad coordinates
lat_hyd = dms_to_deg(17,24,6.5160)     # degrees
lon_hyd = dms_to_deg(78,33,36.6768)     # degrees
h_hyd   = 542

```

```python
#1st question
def geodetic_to_cartesian(lat, lon, h):
    lat = np.radians(lat)
    lon = np.radians(lon)

    N = a / np.sqrt(1 - e2 * np.sin(lat)**2)

    x = (N + h) * np.cos(lat) * np.cos(lon)
    y = (N + h) * np.cos(lat) * np.sin(lon)
    z = ((1 - e2) * N + h) * np.sin(lat)

    return x, y, z
xk, yk, zk = geodetic_to_cartesian(lat_k, lon_k, h_k)
xh, yh, zh = geodetic_to_cartesian(lat_hyd, lon_hyd, h_hyd)
print("Kanpur cartesian coordinates: \n")
print("X:",xk)
print("Y:",xk)
print("Z:",zk)
print("Hyderabad cartesian coordinates: \n")
print("X:",xh)
print("Y:",yh)
print("Z:",zh)
```

```python
def cartesian_to_geodetic(x, y, z, a=a, e2=e2):
    lon = np.arctan2(y, x)
    p = np.sqrt(x**2 + y**2)

    lat = np.arctan2(z, p*(1 - e2))

    for _ in range(5):
        N = a / np.sqrt(1 - e2*np.sin(lat)**2)
        h = p/np.cos(lat) - N
        lat = np.arctan2(z, p*(1 - e2*N/(N+h)))

    return np.degrees(lat), np.degrees(lon), h
lat_K,long_K,H_K = cartesian_to_geodetic(xk,yk,zk,a,e2)
lat_H,long_H,H_H = cartesian_to_geodetic(xh,yh,zh)
print("Kanpur Geodetic: \n")
print("Latitude:",deg_to_dms(lat_K))
print("Longitude:",deg_to_dms(long_K))
print("Height:",H_K)
print("Hyderabad Geodetic: \n")
print("Latitude:",deg_to_dms(lat_H))
print("Longitude:",deg_to_dms(long_H))
print("Height:",H_H)



```

```python
#2nd question
def spherical_to_cartesian(r, lat, lon):
    lat = np.radians(lat)
    lon = np.radians(lon)

    x = r * np.cos(lat) * np.cos(lon)
    y = r * np.cos(lat) * np.sin(lon)
    z = r * np.sin(lat)

    return x, y, z
xk, yk, zk = spherical_to_cartesian(h_k,lat_k,lon_k)
xh,yh,zh = spherical_to_cartesian(h_hyd,lat_hyd,lon_hyd)
print("Spherical (Kanpur):")
print("X =", xk)
print("Y =", yk)
print("Z=", zk)
print("spherical(hyderabad):")
print("X = ",xh)
print("Y =",yh)
print("Z=",zh)




```

```python
def cartesian_to_spherical(x, y, z):
    r = np.sqrt(x**2 + y**2 + z**2)
    lat = np.degrees(np.arcsin(z/r))
    lon = np.degrees(np.arctan2(y, x))
    return r, lat, lon
r_sk, lat_sk, lon_sk = cartesian_to_spherical(xk, yk, zk)
r_sh,lat_sh,lon_sh = cartesian_to_spherical(xh,yh,zh)
#kanpur
print("Spherical (Kanpur):")
print("r =", r_sk)
print("lat =", deg_to_dms(lat_sk))
print("lon =", deg_to_dms(lon_sk))
#hyderabad
print("Spherical (Hyderabad):")
print("r =", r_sh)
print("lat =", deg_to_dms(lat_sh))
print("lon =", deg_to_dms(lon_sh))
```

```python
#3rd question
def geodetic_to_spherical(lat, lon, h):
    x, y, z = geodetic_to_cartesian(lat, lon, h)
    return cartesian_to_spherical(x, y, z)
r_gk, lat_gk, lon_gk = geodetic_to_spherical(lat_sk, lon_sk, r_sk)
r_gh,lat_gh,lon_gh = geodetic_to_spherical(lat_sh,lon_sh,r_sh)
#kanpur
print("Geodetic → Spherical (Kanpur):")
print("r =", r_gk)
print("lat =", deg_to_dms(lat_gk))
print("lon =", deg_to_dms(lon_gk))
#hyderabad
print("Geodetic → Spherical (Hyderabad):")
print("r =", r_gh)
print("lat =", deg_to_dms(lat_gh))
print("lon =", deg_to_dms(lon_gh))
```

```python
#4rth question
def geodetic_to_topocentric(phi_obs, lam_obs, h_obs,
                            phi_sat, lam_sat, h_sat):
    """
    Converts geodetic coordinates of a satellite
    into topocentric ENU coordinates with respect
    to the observer (topocentre)
    """

    # Observer (topocentre) in ECEF
    xo, yo, zo = geodetic_to_cartesian(phi_obs, lam_obs, h_obs)

    # Satellite in ECEF
    xs, ys, zs = geodetic_to_cartesian(phi_sat, lam_sat, h_sat)

    # Displacement vector (Satellite - Observer)
    dx = xs - xo
    dy = ys - yo
    dz = zs - zo

    # ECEF → ENU rotation
    de = -np.sin(np.radians(lam_obs))*dx + np.cos(np.radians(lam_obs))*dy

    dn = (-np.cos(np.radians(lam_obs))*np.sin(np.radians(phi_obs))*dx
          -np.sin(np.radians(lam_obs))*np.sin(np.radians(phi_obs))*dy
          +np.cos(np.radians(phi_obs))*dz)

    du = ( np.cos(np.radians(lam_obs))*np.cos(np.radians(phi_obs))*dx
          +np.sin(np.radians(lam_obs))*np.cos(np.radians(phi_obs))*dy
          +np.sin(np.radians(phi_obs))*dz)

    return de, dn, du
lat_hyd_deg = dms_to_deg(17, 24, 6.5160)
lon_hyd_deg = dms_to_deg(78, 33, 36.6768)
h_hyd   = 542

lat_knp_deg = dms_to_deg(26, 26, 59.7228)
lon_knp_deg = dms_to_deg(80, 19, 54.7356)
h_knp   = 126.630
h_sat   = h_knp + 800000

E, N, U = geodetic_to_topocentric(
    lat_hyd_deg, lon_hyd_deg, h_hyd,
    lat_knp_deg, lon_knp_deg, h_sat
)

print("Topocentric coordinates of the satellite (relative to Hyderabad):")
print(f"East  : {E:.3f} m")
print(f"North : {N:.3f} m")
print(f"Up    : {U:.3f} m")
```

```python
#bonus  question
def topocentric_to_geodetic(E, N, U, lat0, lon0, x0, y0, z0):
    lat0 = np.radians(lat0)
    lon0 = np.radians(lon0)

    dx = (-np.sin(lon0)*E
          -np.sin(lat0)*np.cos(lon0)*N
          +np.cos(lat0)*np.cos(lon0)*U)

    dy = ( np.cos(lon0)*E
          -np.sin(lat0)*np.sin(lon0)*N
          +np.cos(lat0)*np.sin(lon0)*U)

    dz = ( np.cos(lat0)*N
          +np.sin(lat0)*U)

    return x0+dx, y0+dy, z0+dz
E_a, N_a, U_a = -800, 900, 6000

# Get ECEF coordinates of Kanpur observer
xo_kanpur, yo_kanpur, zo_kanpur = geodetic_to_cartesian(lat_k, lon_k, h_k)

# Get ECEF coordinates of Hyderabad observer
# Using lat_hyd_deg, lon_hyd_deg (floats) from previous cells, and h_hyd from kernel state
xo_hyderabad, yo_hyderabad, zo_hyderabad = geodetic_to_cartesian(lat_hyd_deg, lon_hyd_deg, h_hyd)

xak, yak, zak = topocentric_to_geodetic(E_a, N_a, U_a, lat_k, lon_k, xo_kanpur, yo_kanpur, zo_kanpur)
lat_ak, lon_ak, h_ak = cartesian_to_geodetic(xak, yak, zak)

xah,yah,zah = topocentric_to_geodetic(E_a, N_a, U_a, lat_hyd_deg, lon_hyd_deg, xo_hyderabad, yo_hyderabad, zo_hyderabad)
lat_ah,lon_ah,h_ah = cartesian_to_geodetic(xah,yah,zah)
#topocentre is hyderabad
print("Aircraft Geodetic Coordinates of hyderabad:")
print("Latitude :", deg_to_dms(lat_ah))
print("Longitude:", deg_to_dms(lon_ah))
print("Altitude :", h_ah)
```
